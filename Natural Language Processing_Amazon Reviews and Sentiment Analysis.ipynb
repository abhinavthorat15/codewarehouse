```python
# first we convert everything into lower case
amazon.review= amazon.review.str.lower()
```

```python
import string
```

```python
# to check for stopwords we 
from nltk.corpus import stopwords
```

```python
#we can see the list of stopwords in english

stopwords.words("english")
```

```python
def text_process(mess):
    """
    1. remove the punctuation
    2. remove the stopwords
    3.return the list of clean textwords
    """
    nopunc=[char for char in mess if char not in string.punctuation]
    nopunc="".join(nopunc)
    
    return[word for word in nopunc.split() if word not in stopwords.words("english")]
```

```python
#amazon["review"].apply(text_process)
```

```python
# count vectorizer returns the counts of words 
from sklearn.feature_extraction.text import CountVectorizer

bow_transformer= CountVectorizer(analyzer=text_process).fit(amazon["review"])
```

```python
# tell us how many times a specific word has appeared
bow_transformer.vocabulary_
```

```python
# creating a term document matrix
amazon_tdm= bow_transformer.transform(amazon.review)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Sampling of tdm
x_train,x_test, y_train, y_test= train_test_split(amazon_tdm, amazon.label, test_size=.2)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
#logistic 
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
```

```python
logreg.fit(x_train,y_train)
pred_logreg=logreg.predict(x_test)
tab_logreg=confusion_matrix(pred_logreg,y_test)
tab_logreg
```

```python
acc_logreg= tab_logreg.diagonal().sum()/tab_logreg.sum()
acc_logreg
```

# Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
```

```python
#decision Tree
dt.fit(x_train,y_train)
pred_dt=dt.predict(x_test)
tab_dt=confusion_matrix(pred_dt,y_test)
tab_dt
```

```python
acc_dt= tab_dt.diagonal().sum()/tab_dt.sum()
acc_dt
```

# RandomForest

```python
# random forest
rf.fit(x_train,y_train)
pred_rf= rf.predict(x_test)
tab_rf=confusion_matrix(pred_rf, y_test)
tab_rf
```

```python
acc_rf= tab_rf.diagonal().sum()/tab_rf.sum()
acc_rf
```

# sentiment classification


```python
nltk.download("vader_lexicon") # This is a dictionary
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sent=SentimentIntensityAnalyzer()
```

```python
# gives the polarity of a particular review
sent.polarity_scores("this drink was horrible. i had a bad feeling by taking a sip")
```

```python
sent.polarity_scores("this drink was awesome. i had a heavenly feeling by taking a sip")
```

```python
# now suppose if ratings column is not there in amazon dataset
```

```python
# it takes input one sentence at a time
abc= sent.polarity_scores(amazon.iloc[0][1])
```

```python
abc.keys()
```

```python
abc["compound"]
```

```python
# writng a for loop to find all scores
score_com=[]
score_pos=[]
score_neg=[]

for i in range(0,amazon.shape[0]):
    score=sent.polarity_scores(amazon.iloc[i][1])
    score1=score["compound"]
    score_com.append(score1)
    
    score2=score["pos"]
    score_pos.append(score2)
    
    score3=score["neg"]
    score_neg.append(score3)
```

```python
amazon["comp_score"]=score_com
amazon["pos_score"]=score_pos
amazon["neg_score"]=score_neg
```
